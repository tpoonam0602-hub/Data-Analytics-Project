```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
Data = pd.read_csv(r"E-commerce.csv")
df = pd.DataFrame(Data)
print(df)
```

```python
# Show first 5 rows, find duplicate values
print(df.head(5))
print(df.duplicated())
```

```python
# Check Description and Columns Name. 
print(df.describe)
print(df.columns)
```

```python
# Check Data type of data
print(df.dtypes)
```

```python
# Change data type of Time on site
df['Time_on_Site'] = df['Time on Site'].astype('timedelta64[m]')
print(df['Time_on_Site'])
```

```python
# find out null value
print(df.isnull())
print(df.notnull().sum())
```

```python
print(df.columns)
```

```python
# Find out Annual Income
Female_Annual_Income = df.groupby("Gender")['Annual Income'].sum()
print(Female_Annual_Income)
```

```python
# show Average annual income by using bar plot
plt.figure(figsize=(15,8))
sns.barplot(x="Gender", y="Annual Income", data = df, hue= 'Age', palette='viridis')
plt.title("Annual Income")
plt.xticks(rotation=45)
plt.show()
```

```python
# find out Average Annual Incomeaa
Avg_Annual_Income = df.groupby("Gender")['Annual Income'].mean().reset_index()
print(Avg_Annual_Income)
```

```python
# show Average annual income by using bar plot
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Annual Income", data = Avg_Annual_Income, palette="pastel", edgecolor="black")
plt.title("Average Annual Income")
plt.xticks(rotation=45)
plt.show()
```

```python
print(df.columns)
```

```python
# Find out Product Reviews
Product_Review = df.groupby("Product Reviews")["Gender"].sum().sort_index(ascending=False)
print(Product_Review)
```

```python
# find out Rating
Rating = df['Product Reviews']
import json
r=json.loads(Rating[4])

for x in r.items():
    print(x)

```

```python
# Find out Average Time on site
Avg_Time = df["Time_on_Site"].mean()
print(Avg_Time)
```

```python
# Find out Maximum Time browsing
Max_Time = df["Time_on_Site"].max()
print(Max_Time)
```

```python
# Find out Minimum Time browsing
Min_Time = df["Time_on_Site"].min()
print(Min_Time)
```

```python
# Show Avg, Maximum, Minimum time for browsing
plt.figure(figsize=(6,6))
plt.pie(Avg_Time.values, labels=Avg_Time.index, autopct="%1.1f%%", startangle=90)
plt.title("Gender-wise Customer Distribution")
plt.show()
```

```python
# Find out Location wise browsing time
Location_BrowsingTime = df.groupby("Location")["Time_on_Site"].sum()
print(Location_BrowsingTime)
```

```python
# Show Location wise Browsing Time
plt.figure(figsize=(10,5))
sns.scatterplot(x= 'Location', y='Time_on_Site', data=df,  palette='viridis')
plt.title("Location Wise Browsing Time")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    Location_BrowsingTime['Time on Site'],
    labels=Location_BrowsingTime['Location'],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Location-wise Browsing Time Distribution")
plt.show()
```
